```python
import os
import pandas as pd

dataset = "Electronics_5.json"

if os.path.isfile(dataset):
    df = pd.read_json("Electronics_5.json", lines=True)
else:
    url = r"http://deepyeti.ucsd.edu/jianmo/amazon/categoryFilesSmall/Electronics_5.json.gz"
    df = pd.read_json(url, compression='gzip', lines=True)
    

#Remove NA
df.dropna()
#Keep relavent review
df = df[df['reviewText'].str.contains("speaker|headphone|soundsystem|earbud|earphone", na=False, case=False)]
df.shape
df = df[~df.reviewText.str.contains("adapter|cable|screen|book|mount|camera|reader|tablet|radio|walkie talkie|splitter",case=False)]                                                 
df.shape
df = df[~df.asin.str.contains("0972683275|B01H6IEKCW|B01H6IEYIM|B00001P4XH|B00004W3ZP")]                                                 
df.shape
df = df.replace(r'\n\n',' ', regex=True) 
df.head(50)
```

```python
import nltk

sent_text=[]
for review in df.reviewText:    
    sent_text.append(nltk.sent_tokenize(review))
```

```python
import itertools
sent_list = list(itertools.chain.from_iterable(sent_text))
```

```python
len(sent_list)
```

```python
df_new = pd.DataFrame(sent_list,columns=['reviewText'])
df_new.shape
df_new.reviewText[1]
```

```python
df_new = df_new[df_new.reviewText.str.contains("sound|look|design|durable|cost|money|work|pay|cheap|expensive|month|year|base|noise|comfortable|fit|volumn|break|light|easy|quality|size|shape|connect|smooth|buzz|crisp|loud|quiet",case=False)]                                                 
df_new.shape
```

```python
outList = df_new['reviewText'].tolist()
len(outList)
```

```python
import random

random.shuffle(outList)

train = outList[:961392]
test = outList[961392:]
print(len(train))
print(len(test))
```
